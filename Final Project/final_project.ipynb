* 2-) Encoding Multiple Columns and Short Way by using make_column_transformer (Short Way)

```python
df3 = df.copy()
column1 = df3.columns[1]
column2 = df3.columns[2]
column3 = df3.columns[3]
column4 = df3.columns[-1]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

transformer = make_column_transformer( (OneHotEncoder(drop = "first"),[column1,column2,column3,column4]), remainder="passthrough")
print(type(transformer))
transformer
```

```python
transformed = transformer.fit_transform(df3)
transformed_df = pd.DataFrame(transformed, columns = transformer.get_feature_names())
transformed_df
```

* 3-) Use Pandas (Easiest Way) (Dummy Variable)

```python
df
```

```python
y = df["price"]
print("y \n",y)
df = df.drop(columns="price")
display("df",df)
```

```python
column1 = df.columns[1]
column2 = df.columns[2]
column3 = df.columns[3]
```

```python
#Pandas(Shortest Way)
df_dummy = pd.get_dummies(df, prefix=[column1,column2,column3], columns=[column1, column2, column3], drop_first=True)
y_dummy  = pd.get_dummies(y, prefix="price", columns=["price"], drop_first=True)
print("Type of y_dummy:  ",type(y_dummy))
display(y_dummy)
print("value counts of y_dummy\n", y_dummy.value_counts())
```

```python
display(df_dummy)
```

* 4-) One-hot-encoding causes some problems at some times, therefore we may need the label encoding. 

```python
############################### Choose 3 or 4
from sklearn.preprocessing import LabelEncoder

y = LabelEncoder().fit_transform(y)
y = pd.DataFrame(y)
y.value_counts()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler

df_scaled = StandardScaler().fit_transform(df_dummy)
y_scaled = StandardScaler().fit_transform(y_dummy)

print("Type of y_scaled : ", type(y_scaled))
y_scaled = pd.DataFrame(y_scaled)
print("Type of y_scaled : ", type(y_scaled))
display(y_scaled)

print("Type of df_scaled : ", type(df_scaled))
df_scaled = pd.DataFrame(df_scaled)
print("Type of df_scaled : ", type(df_scaled))
display(df_scaled)
```

# Get rid of outliers

```python
from scipy import  stats

stats.zscore(df_scaled, axis=1, ddof=1)
```

```python
z = np.abs(stats.zscore(df_scaled, axis=1, ddof=1))
print("Type of z    : ", type(z))
display(z)
```

```python
print(len(np.where(z > 3)))
print(len(np.where(z > 3)[0]))
print(len(np.where(z > 3)[1]))
```

```python
outliers = list(set(np.where(z > 3)[0]))
print("outliers : ", outliers[0:50])
```

```python
df_cleaned = df_scaled.drop(outliers, axis = 0).reset_index(drop = True)
display(df_cleaned)
#display(df_cleaned.iloc[:, 9:13])
```

```python
y_cleaned = y.drop(outliers, axis = 0).reset_index(drop=True)

display("y_cleaned;", y_cleaned)
print("\n")

print("Value counts of y_cleaned;\n",y_cleaned.value_counts())
```

# Splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_cleaned, y_cleaned, test_size= 0.3, random_state=42)
```

```python
print("Shape of x_train    : ", x_train.shape)
print("Shape of x_test     : ", x_test.shape)
print("Shape of y_train    : ", y_train.shape)
print("Shape of y_test     : ", y_test.shape)
```

# Define Models

```python
# y should be a 1d array, got an array of shape (26276, 4) instead. Therefore we cannot use Logistic Regression
"""from sklearn.linear_model import LogisticRegression

model_logistic = LogisticRegression()
model_logistic.fit(x_train, y_train)

print("Accuracy of train:   ", model_logistic.score(x_train,y_train))
print("Accuracy of train:   ", model_logistic.score(x_test,y_test))
print("Logistic Regression model coeffiecnt:    ", model_logistic.coef_)"""
```

```python
from sklearn.linear_model import LinearRegression

model_linear = LinearRegression()
model_linear.fit(x_train, y_train)

print("Accuracy of train:   ", model_linear.score(x_train,y_train))
print("Accuracy of test:   ", model_linear.score(x_test,y_test), "\n")
#print("Regression model coeffient:  ", model_linear.coef_)
```

```python
from sklearn.linear_model import Ridge

model_ridge = Ridge(alpha = 20)
model_ridge.fit(x_train, y_train)

print("Accuracy of train:   ", model_ridge.score(x_train,y_train))
print("Accuracy of test:   ", model_ridge.score(x_test,y_test), "\n")
#print(f"Ridge model coeffient:  {model_ridge.coef_}")
```

```python
from sklearn.linear_model import Lasso

model_lasso = Lasso(alpha = 0.20)
model_lasso.fit(x_train, y_train)

print("Accuracy of train:   ", model_lasso.score(x_train,y_train))
print("Accuracy of test:   ", model_lasso.score(x_test,y_test), "\n")
#print(f"Lasso model coeffient:  {model_lasso.coef_}")
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth = 4, random_state = 42)
model_clf = clf.fit(x_train, y_train)

print("Accuracy of train:   ", model_clf.score(x_train,y_train))
print("Accuracy of test:   ", model_clf.score(x_test,y_test))
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import Ridge

multi_target_ridge = MultiOutputRegressor(Ridge(random_state=42))
multi_target_ridge.fit(x_train,y_train)

print("Accuracy of train:   ", multi_target_ridge.score(x_train,y_train))
print("Accuracy of test:   ", multi_target_ridge.score(x_test,y_test))
```

```python
multi_target_clf = MultiOutputRegressor(model_clf, n_jobs = 2)
multi_target_clf.fit(x_train,y_train)

print("Accuracy of train:   ", multi_target_clf.score(x_train,y_train))
print("Accuracy of test:   ", multi_target_clf.score(x_test,y_test))
```

```python
##score comparison

#Decision Tree Regressor
print("Decision Tree Regressor")
print("Accuracy of train:   ", model_clf.score(x_train,y_train))
print("Accuracy of test:   ", model_clf.score(x_test,y_test), "\n")

#linear model
print("Simple Linear Model")
print("Simple Train: ", model_linear.score(x_train,y_train))
print("Simple Test: " , model_linear.score(x_test, y_test), "\n")
#print(f"Simple Test: , {model_linear.score(x_test,y_test)} \n")

#Ridge
print("Ridge")
print("Ridge Train: ", model_ridge.score(x_train,y_train))
print("Ridge Test: " , model_ridge.score(x_test, y_test), "\n")

#Lasso
print("Lasso")
print("Lasso Train: ", model_lasso.score(x_train,y_train))
print("Lasso Test: " , model_lasso.score(x_test, y_test), "\n")

#Multi output Ridge
print("Multi output Ridge")
print("Accuracy of train:   ", multi_target_ridge.score(x_train,y_train))
print("Accuracy of test:   ", multi_target_ridge.score(x_test,y_test), "\n")

#Multi output Decision Tree
print("Multi output Decision Tree")
print("Accuracy of train:   ", multi_target_clf.score(x_train,y_train))
print("Accuracy of test:   ", multi_target_clf.score(x_test,y_test))
```
