* Normalize data

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaled_data = MinMaxScaler().fit_transform(df_corr_drop)
scaled_data
```

```python
scaled_data2 = StandardScaler().fit_transform(scaled_data)
scaled_data2
```

# Split your dataset into train and test test (0.7 for train and 0.3 for test).
* Split dataset into train and test set

```python
from sklearn.model_selection import train_test_split, cross_validate

x_train, x_test, y_train, y_test = train_test_split(scaled_data2, yb, test_size = 0.3, random_state = 42)
```

```python
x_train
```

* Import ridge and lasso models from sklearn

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score
```

```python
model_linear = LinearRegression()
model_linear.fit(x_train, y_train)

print(f"Regression model coefficient: {model_linear.coef_}")
```

* Define 3 different alpha values for Ridge and fit them. Print their R^2 sore on both
* train and test.

```python
model_ridge1 = Ridge(alpha = 10)
model_ridge1.fit(x_train, y_train)

print(f"Ridge model coef: {model_ridge1.coef_}")
```

```python
model_ridge2 = Ridge(alpha = 50)
model_ridge2.fit(x_train, y_train)

print(f"Ridge model coef: {model_ridge2.coef_}")
```

```python
model_ridge3 = Ridge(alpha = 100)
model_ridge3.fit(x_train, y_train)

print(f"Ridge model coef: {model_ridge3.coef_}")
```

* Define 3 different alpha values for lasso and fit them. Print their R^2 sore on both
* train and test.

```python
model_lasso1 = Lasso(alpha = 0.1)
model_lasso1.fit(x_train, y_train)

print(f"Lasso model coef: {model_lasso1.coef_}")
```

```python
model_lasso2 = Lasso(alpha = 1)
model_lasso2.fit(x_train, y_train)

print(f"Lasso model coef: {model_lasso2.coef_}")
```

```python
model_lasso3 = Lasso(alpha = 10)
model_lasso3.fit(x_train, y_train)

print(f"Lasso model coef: {model_lasso3.coef_}")
```

```python
##score comparison

#linear model
print("Simple Linear Model")
print("Simple Train: ", model_linear.score(x_train,y_train))
print("Simple Test: " , model_linear.score(x_test, y_test), "\n\n")
#print(f"Simple Test: , {model_linear.score(x_test,y_test)} \n")

#Lasso
print("Lasso 1")
print("Lasso Train: ", model_lasso1.score(x_train,y_train))
print("Lasso Test: " , model_lasso1.score(x_test, y_test), "\n")
print("Lasso 2")
print("Lasso Train: ", model_lasso2.score(x_train,y_train))
print("Lasso Test: " , model_lasso2.score(x_test, y_test), "\n")
print("Lasso 3")
print("Lasso Train: ", model_lasso3.score(x_train,y_train))
print("Lasso Test: " , model_lasso3.score(x_test, y_test), "\n\n")

#Ridge
print("Ridge 1")
print("Ridge Train: ", model_ridge1.score(x_train,y_train))
print("Ridge Test: " , model_ridge1.score(x_test, y_test), "\n")
print("Ridge 2")
print("Ridge Train: ", model_ridge2.score(x_train,y_train))
print("Ridge Test: " , model_ridge2.score(x_test, y_test), "\n")
print("Ridge 3")
print("Ridge Train: ", model_ridge3.score(x_train,y_train))
print("Ridge Test: " , model_ridge3.score(x_test, y_test))
```
